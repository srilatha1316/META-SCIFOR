# SECTION A : DESCRIPTIVE AND STATISTICS

```python
a) Mean:The mean is the average of a set of numbers.

b) Median:The median is the middle value in a sorted list.

c) Mode:The mode is the number that appears most often.
```

```python
import numpy as np
from scipy import stats

data = [2, 4, 4, 6, 8, 10]

mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data, keepdims=True).mode[0]

mean, median, mode
```

```python
*Variance measures the average of the squared differences from the Mean.  
*Standard Deviation is the square root of the variance and represents data spread in the same unit as the data.
```

```python
sample_data = [5, 7, 3, 7, 10, 15]

variance = np.var(sample_data)
std_deviation = np.std(sample_data)

variance, std_deviation
```

```python

print("Types of Variables in Statistics:\n")
print("Quantitative (Numerical): Measured in numbers.")
print("  Example: Height (170 cm)")

print("Qualitative (Categorical): Describes categories or groups.")
print("  Example: Gender (Male/Female)")

print("Discrete Variable: Countable values.")
print("  Example: Number of children (2, 3, etc.)")

print("Continuous Variable: Infinite values within a range.")
print("  Example: Temperature (36.5°C)")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

population = np.random.exponential(scale=2, size=10000)

sample_means = [np.mean(np.random.choice(population, size=30)) for _ in range(1000)]

sns.histplot(sample_means, kde=True)
plt.title("Distribution of Sample Means (Central Limit Theorem)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

```python
5.Independent Variable: The variable we manipulate.
Dependent Variable: The outcome affected by the independent variable.
Example: In a study of how study hours affect test scores:  
- Independent Variable: Hours of study  
- Dependent Variable: Test score
```

```python
6. Measures of Dispersion in Descriptive Statistics:

Range: Difference between max and min values.

Variance: Average squared deviation from the mean.

Standard Deviation: Square root of variance.

Interquartile Range (IQR): Difference between the 75th and 25th percentiles.
```

# Data Visualization using Seaborn & Matplotlib

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
penguins = sns.load_dataset("penguins")
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(tips['total_bill'], bins=20, kde=True)
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.title("Scatterplot of Total Bill vs Tip by Gender")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=tips, x="day", y="total_bill")
plt.title("Boxplot of Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=tips, x="day")
plt.title("Count of Records by Day")
plt.xlabel("Day")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=tips, x="day", y="tip", estimator=np.mean)
plt.title("Average Tip per Day")
plt.xlabel("Day")
plt.ylabel("Average Tip")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(data=iris['petal_length'], fill=True)  # updated from shade=True to fill=True
plt.title("KDE Plot of Petal_Length (Iris Dataset)")
plt.xlabel("Petal_Length")
plt.show()
```

```python
sns.pairplot(data=iris, hue="species")
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(data=tips, x="smoker", y="tip")
plt.title("Violinplot of Tip by Smoker Status")
plt.xlabel("Smoker")
plt.ylabel("Tip")
plt.show()
```

```python
Q1 = tips['total_bill'].quantile(0.25)
Q3 = tips['total_bill'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = tips[(tips['total_bill'] < lower_bound) | (tips['total_bill'] > upper_bound)]
outliers
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(penguins.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap - Penguins Dataset")
plt.show()
```
